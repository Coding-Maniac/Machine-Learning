```python
import matplotlib.pyplot as plt

movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

#barplot
#x-coordiante length of movies [0,1,2,3,4]
plt.bar(range(len(movies)),num_oscars)
plt.title('Favorite movies')
plt.ylabel('Num of oscars')
#label axes with movie names
plt.xticks(range(len(movies)),movies)
plt.show()
```

```python
from collections import Counter
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
# print([min(grade//10 * 10,90) for grade in grades])
histogram = Counter(min(grade//10 * 10,90) for grade in grades)
print(histogram)
print([x for x in histogram.keys()])
plt.bar([x+5 for x in histogram.keys()], #shifting the count (bars) by 5
        histogram.values(),               #Height of the bars
        10,   #size of the bars
        edgecolor=(0,0,0)   #black-color in the edges
       
       )                
plt.axis([-5,105,0,5]) #x-axis from -5 to 105 and y axis from 0 to 5
plt.xticks([10*i for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('# of students')
plt.title('Distribution of students')
plt.show()
"""The third argument to plt.bar specifies the bar width. Here we chose a width of 10, to fill the entire decile. 
We also shifted the bars right by 5, so that, 
for example, the “10” bar (which corresponds to the decile 10–20) 
would have its center at 15 and hence occupy the correct range. 
We also added a black edge to each bar to make them visually distinct."""
```

```python
mentions = [500, 505]
years = [2017, 2018]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")

# if you don't do this, matplotlib will label the x-axis 0, 1
# and then add a +2.013e3 off in the corner (bad matplotlib!)
plt.ticklabel_format(useOffset=False)

# misleading y-axis only shows the part above 500
plt.axis([2016.5, 2018.5, 499, 506])
plt.title("Look at the 'Huge' Increase!")
plt.show()
```

```python
plt.bar(years,mentions,0.8)
plt.xticks(years)
plt.ticklabel_format(useOffset=False)
plt.axis([2016.5,2018.5, 0 ,550])
plt.title('Not so huge anymore')
plt.show()
```

```python
variance     = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x+y for x,y in zip(variance,bias_squared)]
xs = [i for i,_ in enumerate(variance)]

#Multiple plt calls
plt.plot(xs,variance,'g-',label='Variance') #solidgreenline
plt.plot(xs,bias_squared, 'r-',label='Bias_squared') #red dot-dashed
plt.plot(xs,total_error, 'b:' ,label='total_error') #blue dotted

plt.legend(loc=9) #loc=9 means center
plt.xlabel("Model Complexity")
plt.xticks([])
plt.title("The bias variance trade-off")
plt.show()
```

```python
friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

#labeling each point
for label, friend, minute in zip(labels, friends, minutes):
    plt.annotate(label,
                xy=(friend, minute),
                xytext=(5,-5),
                textcoords='offset points')
plt.title("Daily Minutes vs No. of friends")
plt.xlabel("# of friends")
plt.ylabel("Daily minutes spent on site")
plt.show()
```
